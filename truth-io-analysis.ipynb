```python
from claim_processer import process_claims
```

```python
statements = [
    "The Eiffel Tower is located in Paris, France.",
    "The Amazon River is the longest river in the world.",
    "Penguins are flightless birds found primarily in the Southern Hemisphere, with most species living in Antarctica.",
    "The Great Wall of China was built entirely during the reign of the Ming Dynasty.",
    "The human brain contains approximately 86 billion neurons, each capable of forming thousands of synaptic connections, facilitating complex cognitive functions.",
    "Sharks are mammals because they give birth to live young and produce milk for their offspring.",
    "The capital of Japan is Tokyo, and it has been the capital since 500 BCE.",
    "Solar panels convert sunlight into electricity through photovoltaic cells, but they cannot generate power during cloudy weather.",
    "Mount Everest, the highest mountain on Earth, is 8,849 meters tall.",
    "Marie Curie won a Nobel Prize in Physics and another in Biology for her research on radioactivity."
]
```

```python
token_counts = []
for statement in statements:
    _, count = process_claims(statement)
    token_counts.append(count)
```

```python
token_counts
```

```python
# Calculate average total token usage across all tests
total_tokens = [sum(count.values()) for count in token_counts]
average_tokens = sum(total_tokens) / len(total_tokens)
print(f"Average total tokens per test: {average_tokens:.2f}")
```
